## Change column values to more meaningful names

```python
column_mapping = {'Unnamed: 0': 'Index','track_id': 'Song ID','artists': 'Artists','album_name': 'Album Name','track_name': 'Song Name'
                  ,'popularity': 'Popularity','duration_ms': 'Duration (ms)','explicit': 'Explicit','danceability': 'Danceability','energy': 'Energy',
                  'key': 'Key','loudness':'Loudness','mode': 'Mode','speechiness': 'Speechiness','acousticness': 'Acousticness','instrumentalness': 'Instrumentalness',
                  'liveness': 'Liveness','valence': 'Valence','tempo': 'Tempo','time_signature': 'Time Signature','track_genre': 'Genre'}
spotify_DF = spotify_DF.rename(columns=column_mapping)
spotify_DF.columns
```

## The meaning of the column values

1. `Song ID`: The Spotify ID for the song
2. `Artists`: The artists' names who performed the song
3. `Album Name`: The album name in which the song appears
4. `Song Name`: Name of the song
5. `Popularity`: The popularity of a song is a value between 0 and 100, with 100 being the most popular
6. `Duration (ms)`: The song length in milliseconds
7. `Explicit`: Whether or not the song has explicit lyrics (true = yes it does; false = no it does not OR unknown) 
8. `Danceability`: Describes how suitable a track is for dancing,A value of 0.0 is least danceable and 1.0 is most danceable
9. `Energy`: The energy of the song, a value ranging from 0 to 1
10. `Key`: The key the song is in
11. `Loudness`: The overall loudness of a track in decibels (dB)
12. `Mode`:  The mode (major or minor) of the song
13. `Speechiness`:  The presence of spoken words in the song, ranging from 0 to 1
14. `Acousticness`: The likelihood that the song is acoustic, ranging from 0 to 1
15. `Instrumentalness`: The likelihood that the song does not contain vocals, ranging from 0 to 1
16. `Liveness`: The probability that the song was recorded live (capturing audience presence), ranging from 0 to 1
17. `Valence`: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track
18. `Tempo`: The overall estimated tempo of a track in beats per minute (BPM)
19. `Time Signature`: The estimated time signature of the song, ranging from 3 to 7, representing the rhythmic time signature
20. `Genre`: The genre in which the song belongs

## Let's look at genres and their valence rating:

```python
valence_genre = spotify_DF[['Genre','Valence']]
valence_genre.groupby('Genre')['Valence'].mean().reset_index()\
                        .sort_values('Valence',ascending=False).head(11)
```

#### After looking at the top of the list, we didn't find anything surprising, and we decided to look at the bottom of the list instead.

```python
valence_genre.groupby('Genre')['Valence'].mean().reset_index()\
                        .sort_values('Valence',ascending=False).tail(11)
```

```python
avg_valence_Genre = valence_genre.groupby('Genre')[['Valence']].mean().reset_index()
avg_valence_Genre_sorted = avg_valence_Genre.sort_values('Valence', ascending=False).tail(11)

colors = plt.cm.viridis(np.linspace(0, 1, len(avg_valence_Genre_sorted )))
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(avg_valence_Genre_sorted ['Genre'], avg_valence_Genre_sorted ['Valence'], color=colors)
ax.set_xlabel('Average Valence')
ax.set_ylabel('Genre')
ax.set_title('Average Valence by Genre (Bottom 10)')
plt.show()
```

#### Observation 1:
##### The graph displays the 10 genres out of all genres with the lowest valence values. We can see genres such as "trance," "dubstep," and "new-age" among them. According to their definition, they are characterized as creators of uplifting and euphoric emotions, which contradicts and goes against logic.

## Let's look at the valence for each popularity level:

```python
columns_of_interest = ['Popularity', 'Valence']
popularity_avg = spotify_DF.groupby('Popularity')[columns_of_interest].mean()
plt.plot(popularity_avg.index, popularity_avg['Valence'], marker='o')
plt.xlabel('Popularity')
plt.ylabel('Valence')
plt.title('Average Valence for Each Popularity Level')

plt.show()
```

#### In the majority of the graph, we can see that the valence rating is relatively moderate (between 0.4-0.6), but towards the end of the graph, there are more extreme fluctuations. What surprised us to see is that, on average, less positive songs are more popular.

```python
lives = spotify_DF[spotify_DF['Song Name'].str.contains('- live', case=False)]
lives_genre= lives[['Liveness','Song Name','Genre']].sort_values('Liveness')
last_five_genres = lives_genre.groupby('Genre')[['Liveness']].mean().sort_values('Liveness',ascending=False)
last_five_genres_tail = last_five_genres.tail(10).reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=last_five_genres_tail['Genre'], y=last_five_genres_tail['Liveness'])
plt.xlabel('Genre')
plt.ylabel('Liveness')
plt.title('Liveness for Last Five Genres')
plt.xticks(rotation=45)
plt.show()
```

#### Observation 2:
##### We were surprised to discover that there are a great number of songs that despite being labeled as "live," have an extremely low "liveness" rating. After consulting with an expert in the field of music, the explanation appears to be due to a discrepancy in the recording method of the songs uploaded to the application (microphone recording versus speaker recording).

## Correlations - patterns in the data:

```python
relevent_corr = spotify_DF[['Artists', 'Album Name', 'Song Name',
        'Popularity', 'Explicit', 'Danceability', 'Energy',
        'Loudness', 'Speechiness', 'Acousticness',
        'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Time Signature',
        'Genre']]
corr_table = relevent_corr.corr(method="pearson")
corr_table
```

```python
plt.figure(figsize=(16,4))
sns.heatmap(corr_table, annot=True, center=0, cmap='PiYG')
plt.title("Correlation Heatmap between variables")
plt.show()
```

```python
correlation_matrix = spotify_DF.corr().abs().unstack()
highest_correlation = correlation_matrix[(correlation_matrix < 1) & (correlation_matrix != -1)].max()
feature1, feature2 = correlation_matrix[(correlation_matrix == highest_correlation)].idxmax()
print(f"The highest correlation (excluding 1 or -1) is between '{feature1}' and '{feature2}', with a correlation value of {highest_correlation}.")
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))

sns.regplot(x=spotify_DF['Energy'], y=spotify_DF['Loudness'], scatter_kws={"s": 0.5}, line_kws={"color": "lightblue"})
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Linear Regression of Energy and Loudness')

plt.show()
```

#### We noticed that as the energy increases, the loudness level also increases. Additionally, the tempo (beats per minute) remains unchanged. We expected that if the volume and energy increase, the tempo would necessarily increase as well, which is not the case, the correlation is 0.21(tempo-loudness) and 0.25 (tempo-energy).

## Preference for a song with offensive lyrics versus a song without offensive lyrics based on popularity:

```python
avg_popularity = spotify_DF.groupby(['Explicit'])['Popularity'].sum().to_frame()

fig, ax3 = plt.subplots(figsize=(6,5))
ax3.bar(avg_popularity.index, avg_popularity['Popularity'], color=['#FF9999', '#66B2FF'])
ax3.set_xticks(avg_popularity.index)
ax3.set_xticklabels(['Non-offensive text', 'Offensive text'])
ax3.set_title('Offensive vs. non-offensive popular songs')

plt.tight_layout()
plt.show()
```

####


##### The code compares the total popularity levels of songs with offensive text and songs with non-offensive text. The graph illustrates the total **sums** of popularity for the two categories. Each column represents one category, and its height represents the sum of popularity for songs in that category. The graph shows that songs with non-offensive text have a higher total popularity sum than songs with offensive text.

```python
explicit_count = spotify_DF['Explicit'].value_counts().to_frame()
explicit_count
```

#### The code calculates the number of values in the 'Explicit' column. There are two possible values: True=Offensive text and False=Non-offensive text. This allows us to understand the distribution of the values and their relationship.

```python
explicit_count = spotify_DF['Explicit'].value_counts()
total_popularity = spotify_DF.groupby(['Explicit'])['Popularity'].mean().to_frame()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.pie(explicit_count.values, labels=['Non-offensive text', 'Offensive text'], colors=['#FF9999', '#66B2FF'],
        autopct='%1.1f%%', startangle=90)
ax1.axis('equal') 
ax1.set_title('Offensive songs V.S non-offensive songs in percentages')

ax2.bar(total_popularity.index, total_popularity['Popularity'], color=['#FF9999', '#66B2FF'])
ax2.set_xticks(total_popularity.index)
ax2.set_xticklabels(['Non-offensive text', 'Offensive text'])
ax2.set_title('Total avg popularity of offensive and non-offensive songs')

plt.tight_layout()
plt.show()
```
